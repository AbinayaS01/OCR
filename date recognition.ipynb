```python
import re
import cv2
import pytesseract
from pytesseract import Output
```

```python
img  = cv2.imread("date.png")
```

```python
from PIL import Image
```

```python
img1 = Image.open("date.png")
```

```python
d = pytesseract.image_to_data(img,output_type=Output.DICT)
```

```python
keys=list(d.keys())
```

```python
data_pattern='^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|[012])/(19|20)\d\d$'
```

```python
n_boxes = len(d['text'])
for i in range(n_boxes):
    if int(d['conf'][i])>60:
        if re.match(data_pattern,d['text'][i]):
            (x,y,w,h)=(d['left'][i],d['top'][i],d['width'][i],d['height'][i])
            img = cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

cv2.imshow('img',img)
cv2.waitKey(0)
```
